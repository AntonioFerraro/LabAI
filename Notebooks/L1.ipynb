## Call model on personal space

### api url is composed by username-spacename.hf.space/endpoint (the post endpoint you defined in app.py)
### username-spacename should be lowercase

```python
import requests

API_URL = "YOUR_URL"

payload = {
    "prompt": """
You are a classifier.

Task: Detect discrimination in a sentence.

Rules:
- If the sentence is discriminatory → answer YES
- If not → answer NO
- Respond with ONLY one word. Don't add an explanation

Sentence: "neapolitans cannot have a job"
Answer:
""",
    "temperature": 0,
    "max_tokens": 5
}

response = requests.post(API_URL, json=payload)

print(response.json()["response"])
```

### You can also call inference api with bigger models, but be careful about usage quota

https://huggingface.co/docs/inference-providers/pricing#free-credits-to-get-started

```python
import os
from openai import OpenAI

client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key="CREATED_READ_ACCESS_TOKEN",
)

completion = client.chat.completions.create(
    model="moonshotai/Kimi-K2-Instruct-0905",
    messages=[
        {
            "role": "user",
            "content": """
You are a classifier.

Task: Detect discrimination in a sentence.

Rules:
- If the sentence is discriminatory → answer YES
- If not → answer NO
- Respond with ONLY one word.

Sentence: "neapolitans cannot have a job"
Answer:
"""
        }
    ],
    temperature=0,
    max_tokens=5
)

print(completion.choices[0].message.content)
```
