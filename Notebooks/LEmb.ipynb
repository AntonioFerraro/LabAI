```python
import requests
import numpy as np
```

```python
def get_embedding(text):
    url = "https://bard53-mmbert.hf.space/embed"

    response = requests.post(
        url,
        json={"text": text},
        timeout=60,
        verify=False
    )

    if response.status_code != 200:
        raise Exception(f"Error: {response.text}")

    return response.json()

def cosine_similarity_np(vec1, vec2):
    v1 = np.array(vec1)
    v2 = np.array(vec2)

    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
emb1 = get_embedding('Cat is on the table')['embedding']
emb2 = get_embedding('Cat is singing on the table')['embedding']

cosine_similarity_np(emb1, emb2)
```

```python
emb1 = get_embedding('Cat is on the table')['embedding']
emb2 = get_embedding('Cat is running on a hill')['embedding']

cosine_similarity_np(emb1, emb2)
```

```python
emb1 = get_embedding('Cat water dancing')['embedding']
emb2 = get_embedding('Cat swimming')['embedding']

cosine_similarity_np(emb1, emb2)
```
