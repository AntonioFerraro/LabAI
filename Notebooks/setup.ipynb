### Miniforge Installaltion

download from: https://conda-forge.org/download/

based on your OS

verify:

- conda --version
- mamba --version

### Postgre setup

on miniforge terminal:

conda create -n pgvector-env -c conda-forge postgresql pgvector psycopg sqlalchemy pandas jupyterlab -y+

activate env: conda activate pgvector-env

verify:

- postgres --version
- psql --version

### Create Postgre Cluster

mkdir %USERPROFILE%\pgvector-conda-data

check: dir %USERPROFILE%\pgvector-conda-data

init db: initdb -D "%USERPROFILE%\pgvector-conda-data" -U postgres -A scram-sha-256 -W

to change door: go to %USERPROFILE%\pgvector-conda-data\postgresql.conf 

and change to: (delete # if there)

- port = 5433
- listen_addresses = 'localhost'

in miniforge prompt start server with: pg_ctl -D "%USERPROFILE%\pgvector-conda-data" -l "%USERPROFILE%\pgvector-conda-data\server.log" start

### Create DB

open psql (after postgre download)

psql -h localhost -p 5433 -U postgres -d postgres

use password inserted before

Steps:

- CREATE DATABASE vectordb;
- \c vectordb
- CREATE EXTENSION vector;
